```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline 
import numpy as np
```

## read in all data

```python
apparel = pd.read_excel('apparel.xlsx', header=10)
food = pd.read_excel('food.xlsx', header=10)
housing = pd.read_excel('housing.xlsx', header=10)
trans = pd.read_excel('transportation.xlsx', header=10)
other = pd.read_excel('other.xlsx', header=10)
medcare = pd.read_excel('medcare.xlsx', header=10)
```

```python
apparel.head()
```

## quick graphs

```python
fig, ax = plt.subplots()
ax.plot(apparel['Year'],apparel['Annual'], color='r')
ax.plot(food['Year'], food['Annual'], color='b')
ax.plot(trans['Year'], trans['Annual'], color='black')
ax.plot(medcare['Year'], medcare['Annual'], color='y')
ax.plot(other['Year'], other['Annual'], color ='orange')
ax.plot(housing['Year'], housing['Annual'], color='g')

ax.axis([1980, 2015, 0,450])

ax.axvspan(2007, 2009, alpha=0.15, color='gray')
ax.axvspan(1990, 1991, alpha=0.15, color='gray')
ax.axvspan(1980, 1982, alpha=0.15, color='gray')
```

## normalize 

```python
def normalize(var): 
    new_var = (var - var.min())/(var.max()-var.min())
    return new_var
```

```python
apparel['Annual_norm'] = normalize(apparel['Annual'])
food['Annual_norm'] = normalize(food['Annual'])
trans['Annual_norm'] = normalize(trans['Annual'])
medcare['Annual_norm'] = normalize(medcare['Annual'])
other['Annual_norm'] = normalize(other['Annual'])
housing['Annual_norm'] = normalize(housing['Annual'])
```

```python

fig, ax = plt.subplots()
ax.plot(apparel['Year'],apparel['Annual_norm'], color='r')
ax.plot(food['Year'], food['Annual_norm'], color='b')
ax.plot(trans['Year'], trans['Annual_norm'], color='black')
ax.plot(medcare['Year'], medcare['Annual_norm'], color='y')
ax.plot(other['Year'], other['Annual_norm'], color ='orange')
ax.plot(housing['Year'], housing['Annual_norm'], color='g')

ax.axis([1980, 2015, 0,1])

ax.axvspan(2007, 2009, alpha=0.15, color='gray')
ax.axvspan(1990, 1991, alpha=0.15, color='gray')
ax.axvspan(1980, 1982, alpha=0.15, color='gray')
```

## percent change

```python
apparel['Annual_pct_change'] = apparel['Annual'].pct_change(periods=1)
food['Annual_pct_change'] = food['Annual'].pct_change(periods=1)
trans['Annual_pct_change'] = trans['Annual'].pct_change(periods=1)
medcare['Annual_pct_change'] = medcare['Annual'].pct_change(periods=1)
housing['Annual_pct_change'] = housing['Annual'].pct_change(periods=1)
other['Annual_pct_change'] = other['Annual'].pct_change(periods=1)
```

```python

fig, ax = plt.subplots()
ax.plot(apparel['Year'],apparel['Annual_pct_change'], color='r')
ax.plot(food['Year'], food['Annual_pct_change'], color='b')
ax.plot(trans['Year'], trans['Annual_pct_change'], color='black')
ax.plot(medcare['Year'], medcare['Annual_pct_change'], color='y')
ax.plot(other['Year'], other['Annual_pct_change'], color ='orange')
ax.plot(housing['Year'], housing['Annual_pct_change'], color='g')

ax.axis([1980, 2015, -.10,.15])

ax.axvspan(2007, 2009, alpha=0.15, color='gray')
ax.axvspan(1990, 1991, alpha=0.15, color='gray')
ax.axvspan(1980, 1982, alpha=0.15, color='gray')
```
