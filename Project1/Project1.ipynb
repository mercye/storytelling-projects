```python
import pandas as pd
```

```python
# before reading in, discarded any commitment amount under 100 dollars
df = pd.read_csv('aiddata2-1_thin_edit.csv')
```

```python
df.head()
```

```python
df2 = df[df['donor']=='United States']
df3 = df2.groupby(['recipient','year','coalesced_purpose_code']).sum()
```

```python
df2.head()
```

```python
df3.head()
```

```python
del df3['aiddata_id']
```

```python
df3.columns = ['year_purpose_total']
df3.head(1000)
```

```python
df4 = df2.merge(df3, how='right', left_on=['recipient','year','coalesced_purpose_code'], right_index=True)
```

```python
df4.head()
```

```python
df4.drop_duplicates(['year', 'recipient', 'coalesced_purpose_code'], keep='first', inplace=True)
```

```python
df5 = df4.groupby(['recipient','year']).sum()
```

```python
del [df5['aiddata_id'], df5['commitment_amount_usd_constant'], df5['coalesced_purpose_code']]
```

```python
df5.head()
```

```python
df5.columns = ['year_total']
```

```python
df6 = df4.merge(df5, how='right', left_on = ['recipient','year'] , right_index=True )
```

```python
del [df6['commitment_amount_usd_constant'], df6['aiddata_id']]
```

```python
df6.head()
```

```python
df7 = df6.sort_values(['year', 'year_total'], ascending=False)
```

```python
df8 = df7[df7['year'] == 2010]
```

```python
df9 = df8[df8['recipient'] != 'Bilateral, unspecified']
```

```python
df10 = df9.sort_values('year_total', ascending=False)
```

```python
bins = [11000, 11999, 13999, 14999, 20999,22999,23999,30999,51999,69999,89999,100000]
group_names = ['Education','Health','Water','Civil Society', 'Trans/Comm', 'Energy', 'Banking/Business', 'Agriculture/Env', 'Food', 'Emergeny Disaster Relief', 'Administrative']
categories = pd.cut(df10['coalesced_purpose_code'], bins, labels=group_names)
df10['categories'] = pd.cut(df10['coalesced_purpose_code'], bins, labels=group_names)
df10.head()
```

```python
df10.drop_duplicates(['year', 'recipient', 'categories'], keep='first', inplace=True)
```

```python
df10.to_csv('aid.csv')
```
